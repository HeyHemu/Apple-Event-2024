```python
# Python program to count word frequencies with extended stop words removal

from collections import Counter
import matplotlib.pyplot as plt
import string

# Extended list of stop words and fillers
stop_words = set(STOPWORDS)

# Read the .txt file
with open(r"/content/drive/MyDrive/Colab Notebooks/apple event.txt", encoding="utf-8") as file:
      text = file.read()

# Tokenize and clean the text
words = text.lower().translate(str.maketrans("", "", string.punctuation)).split()
cleaned_words = [word for word in words if word not in stop_words and len(word) > 1]

# Count word frequencies
word_freq = Counter(cleaned_words)

# Get the 20 most common words
most_common = word_freq.most_common(10)

# Prepare data for plotting
words, counts = zip(*most_common)

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.title("Top 10 Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()

# Print the word frequencies
print("Word Frequencies:")
for word, count in most_common:
    print(f"{word}: {count}")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords data
import nltk
nltk.download('stopwords')
nltk.download('punkt')

def generate_wordcloud(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

    # Join the filtered tokens back into a string
    filtered_text = ' '.join(filtered_tokens)

    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=600, background_color='white').generate(filtered_text)

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()

# Read the text from the file
file_path = "/content/drive/MyDrive/Colab Notebooks/apple event.txt"
with open(file_path, encoding="utf-8") as file:
    text = file.read()

# Generate the word cloud
generate_wordcloud(text)
```
